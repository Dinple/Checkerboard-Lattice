```python
import os
import numpy as np
from CBLattice import *
```

```python
# Create a Chekcerboard Lattice
cb_lattice = CheckerBoard(n_dim_=10, n_act_=2)
cb_lattice.print_lattice()
```

```python
# opena a file to write the smt variables
f = open("./inputSMT/problem_01.smt2", "w")

# write the smt variables
cb_lattice.preface(f)

cb_lattice.collect_var()

cb_lattice.collect_abstract_var()

cb_lattice.declare_var(f)

cb_lattice.init_var(f)

cb_lattice.assert_matrix_const(f)

cb_lattice.assert_time_abstract(f)

cb_lattice.assert_entropy_abstract(f)

# cb_lattice.assert_swap_op(f)

cb_lattice.assert_final_obj(f)

cb_lattice.assert_obj_func(f)

cb_lattice.conclude(f)

f.close()
```

```python
# create the solution file
f = open("./z3/problem_01.z3", "w")

# call z3 to solve the problem
os.system("z3 ./inputSMT/problem_01.smt2 > ./z3/problem_01.z3")

# read the solution
f = open("./z3/problem_01.z3", "r")
lines = f.readlines()
f.close()
```

```python
# if the solution is SAT, then print the solution
if lines[0] == "sat\n":
    print("SAT")
    # print all the variables
    cb_lattice.read_res(lines)
    # cb_lattice.print_true_qbit()
    # cb_lattice.print_true_qbit_map()
    # cb_lattice.print_true_m()
    # cb_lattice.print_true_QL()
    # cb_lattice.print_true_CL()
    # cb_lattice.print_true_RL()
    # cb_lattice.print_true_time()
    # cb_lattice.print_true_obj()
    cb_lattice.view_solution()

else:
    print("UNSAT")
    
```
